## 3) Treatment of outliers for the relevant dataframe.

##### Treating each column individually and doing univariate analysis on it. 

```python
# We can do this treatment using Inter Quartile Range Outlier Detection Method.
# This variable denotes the total income of a person.
income = relevant_df['AMT_INCOME_TOTAL']
Q1 = income.quantile(0.25)
Q3 = income.quantile(0.75)
Range = Q3 - Q1
df_income_outlier =  income[~((income < (Q1 - 1.5 * Range)) |(income > (Q3 + 1.5 * Range)))]
df_income_outlier
```

##### Age Outlier

```python
Age = relevant_df['DAYS_BIRTH']* -1 
S1 = Age.quantile(0.25)
S3 = Age.quantile(0.75)
Range1 = S3 - S1

df_outlier_age =  Age[~((Age < (S1 - 1.5 * Range1)) | (Age > (S3 + 1.5 * Range1)))]
df_outlier_age
```

#### Price

```python
Goods_price = relevant_df['AMT_GOODS_PRICE']
R1 = Goods_price.quantile(0.25)
R3 = Goods_price.quantile(0.75)
Range2  = R3 - R1

df_outlier_price =  Goods_price[~((Goods_price < (R1 - 1.5 * Range2)) | (Goods_price > (R3 + 1.5 * Range2)))]
df_outlier_price
```
